```python

def get_path_length(path, G):
    path_length = 0
    
    for i in range(len(path) - 1):
        v1 = path[i]
        v2 = path[i + 1]
        
        if not G.has_edge(v1, v2):
            
            return float('inf')
       
        path_length += G[v1][v2]["weight"]
    return path_length


def node_permutations(G, init_node):
    nodes = list(G.nodes())
    nodes.remove(init_node)
    return [[init_node] + list(a_tuple) for a_tuple in permutations(nodes)]


def TSP_BruteForce(G, init_node):
    min_path = None
    min_path_length = float('inf')
    
    for path in node_permutations(G, init_node):
        
        path_length = get_path_length(path, G)
       
        if path_length < min_path_length:
            
            min_path = path
            min_path_length = path_length
   
    return min_path, min_path_length
```

```python
import networkx as nx
import matplotlib.pyplot as plt


G = nx.Graph()
G.add_weighted_edges_from([(1, 3, 6), (1, 4, 8), (1, 5, 10), (2, 3, 5), (2, 4, 7), (2, 5, 9)])


pos = nx.spring_layout(G) 


nx.draw_networkx_nodes(G, pos, node_size=700)
nx.draw_networkx_labels(G, pos, font_size=20, font_family='sans-serif')


edges = [(u, v) for (u, v, d) in G.edges(data=True)]
nx.draw_networkx_edges(G, pos, edgelist=edges)


edge_labels = {(u, v): d['weight'] for u, v, d in G.edges(data=True)}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)


plt.show()
```
